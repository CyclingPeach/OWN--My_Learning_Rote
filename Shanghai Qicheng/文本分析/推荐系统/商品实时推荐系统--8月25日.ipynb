````python
'''
    根据评分对用户和商品进行排序
'''
id_sort = df.groupby(by='sku_id')['score'].count().sort_values(ascending=False)
id_sort
````

用**```quantiles()```**函数返回分位情况

在统计学里经常提到分位数的概念，下面我们通过一个例子来简单解释一下。

假设有一千名学生参加了某次考试，
学生A得了75分，排名603，603/1000＝60.3%
学生B得了94分，排名28，28/1000=2.8%
此时，A大约在60.3%的位置上，而B大约在2.8%的位置上。即在60.3%的位置上约75分, 2.8%的位置上约94分。

对应四分位数的就很好解释了，分别在25%, 50%, 75%位置上的数。假设考生甲乙丙丁考试成绩分别为80,71,61，对应的名次分别为250,500,750名，那么对应的四分位数分别就为80,71,61。

````python
import matplotlib.pyplot as plt
````

````python
np.arange(0, 10, 1)
# 包含0，不包含10
````

````python
'''
    返回商品的评价数量的四分位数
'''
quantiles = id_sort.quantile(np.arange(0,1.01,0.01), interpolation='higher')
plt.figure(figsize=(10,5))
plt.title("Quantiles and their Values")
quantiles.plot()
# quantiles with 0.05 difference
plt.scatter(x=quantiles.index[::5], y=quantiles.values[::5], c='orange', label="quantiles with 0.05 intervals")
# quantiles with 0.25 difference
plt.scatter(x=quantiles.index[::25], y=quantiles.values[::25], c='m', label = "quantiles with 0.25 intervals")
````

````python
pd.DataFrame(quantiles)
````

## 基于商品流行度实现商品推荐

### ```filter()```函数

filter() 函数用于**过滤序列**，过滤掉不符合条件的元素，返回由符合条件元素组成的新列表。

该接收两个参数，第一个为函数，第二个为序列，序列的每个元素作为参数传递给函数进行判断，然后返回 True 或 False，最后将返回 True 的元素放到新列表中。

````python
import matplotlib.pyplot as plt
````

````python
'''
    对商品进行排序
'''
new_df = df.groupby('sku_id').filter(lambda x:x['score'].count() >=50)
id_sort

fig = plt.figure(figsize=plt.figaspect(0.5))    # 设置长宽比例
ax = plt.gca()    # 
plt.plot(id_sort.values)
plt.title('# RATINGS per Product')   # 每个商品的评价
plt.xlabel('Product')    # 商品
plt.ylabel('id_sort')
ax.set_xticklabels([])
plt.show()

# 这里的plt.gca()可以参见：https://www.cnblogs.com/ymjyqsx/p/7390288.html
````

## 注意这里的分组和排序
### 同时还要注意如何自定义列名

````python
'''
    计算出每个商品的平均评分，然后再按平均分进行降序排序
    默认为升序排序，降序需要设置ascending=False
'''
pd.DataFrame(new_df.groupby('sku_id')['score'].mean().sort_values(ascending=False))
````

````python
'''
    计算每个商品被评分次数的排名
'''
pd.DataFrame(new_df.groupby('sku_id')['score'].count().sort_values(ascending=False))
````

````python
'''
合并到一起
'''

ratings_mean_count = pd.DataFrame(new_df.groupby('sku_id')['score'].mean())

'''新增列，并设置列名'''
ratings_mean_count['score_counts'] = pd.DataFrame(new_df.groupby('sku_id')['score'].count())
ratings_mean_count
````

## 结果可视化

````python
plt.figure(figsize=(10, 5))    # 设置图像大小
#plt.rcParams['patch.force_edgecolor'] = True
ratings_mean_count['score_counts'].hist(bins=50)    # hist() 绘制直方图，bins 绘制直方图的桶数
````

````python
'''
    每个评分的用户数量，并排序
'''
ratings_user_count = pd.DataFrame(df.groupby('score')['_id'].count().sort_values(ascending=False))
ratings_user_count
````

````python
'''
    使用直方图绘制出每个评分级别的用户数量
'''
plt.figure(figsize=(10,5))
plt.rcParams['patch.force_edgecolor'] = True
ratings_user_count['_id'].hist(bins=10)
````

````python
'''
    使用散点图和直方图绘制每个级别的用户数量数据分布
'''
import seaborn as sns

plt.figure(figsize=(10,5))
plt.rcParams['patch.force_edgecolor'] = True
sns.jointplot(x='score', y='score_counts', data=ratings_mean_count, alpha=0.4)
````

## 基于协同过滤算法实现商品推荐
已经针对商品的流行度实现了推荐，但是说服力还是不足以满足现实的情况，进而使用目前比较流行的协同过滤的方法实现推荐系统。

### 创建商品和用户的关系表格

首先选取样本，然后再使用pivot_table()创建**用户和商品的关系**表格

````python
new_df1 = new_df.head(1000)
ratings_matrix = new_df1.pivot_table(values='score', index='_id', columns='sku_id', fill_value=0)
ratings_matrix
````

````python
ratings_matrix.shape
````

````python
X = ratings_matrix.T
X
````

## 分解表格
使用**SVD**算法对表格进行降维，得到基于商品的几个重要特征数据

````python
from sklearn.decomposition import TruncatedSVD # 导入 SVD 算法
SVD = TruncatedSVD(n_components=10)# 构建一个 SVD 模型，其作用是将特征数目即列数组合为 10 个重要组合特征

decomposed_matrix = SVD.fit_transform(X)# 对表格进行转化
decomposed_matrix.shape# 查看转化后的表格大小
````

````python
pd.DataFrame(decomposed_matrix)
````

### 构建相关系数矩阵
算出商品间的相识性，用以做基于物品的推荐系统

````python
correlation_matrix = np.corrcoef(decomposed_matrix)    # corrcoef()：算出数据之间的相关系数
print(correlation_matrix.shape)
pd.DataFrame(correlation_matrix)
````
